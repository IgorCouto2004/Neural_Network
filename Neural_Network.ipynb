```python
import pandas as pd
```

```python
dados = pd.read_excel('churn.xlsx')
```

```python
dados.head()
```

```python
dados = dados.drop(columns=['customerID'])
dados.head()
```

```python
dados = dados.dropna()
```

```python
dados.isnull().any()
```

```python
dados.dtypes
```

```python
dados.TotalCharges = pd.to_numeric(dados.TotalCharges, errors='coerce')
```

```python
dados = dados.dropna()
```

```python
dados.head()
```

```python
cat_vars = ['PhoneService', 'Contract', 'PaperlessBilling', 'PaymentMethod']
num_vars = ['tenure', 'MonthlyCharges', 'TotalCharges']
modeling_vars = cat_vars + num_vars
```

```python
X = dados.filter(modeling_vars).copy()
y = dados['Churn'].copy()
y = pd.DataFrame(y)
```

```python
#Churn não está em formato binário
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
#Agora está :)
```

```python
y
```

```python
X = pd.get_dummies(X)
```

```python
dados['Churn'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

clf = MLPClassifier(hidden_layer_sizes=(100, 50, 20),activation='logistic')
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(X_test_scaled)
print(f'Acurácia do modelo: {accuracy_score(y_test, y_pred)}')
```
